# Descripcion de las columnas
'Unnamed: 0': Esta columna parece ser un índice o un identificador único para cada registro en el conjunto de datos. Puede ser ignorada o eliminada si no aporta información relevante.

'fecha': Esta columna registra la fecha del accidente de vuelo.

'HORA declarada': Esta columna registra la hora declarada del accidente de vuelo. Presumiblemente, es la hora en que ocurrió el accidente o la hora estimada proporcionada oficialmente.

'Ruta': Esta columna indica la ruta planificada o programada del vuelo.

'OperadOR': Esta columna representa la aerolínea o la compañía que operaba el vuelo.

'flight_no': Esta columna contiene el número de vuelo asociado con el accidente.

'route': Esta columna puede ser similar a la columna 'Ruta' mencionada anteriormente. Podría ser una duplicación o tener información adicional.
actualizacion-> tiene informacion sobre el tipo de vuelo que se hizo. 

'ac_type': Esta columna indica el tipo de aeronave involucrada en el accidente.

'registration': Esta columna registra el registro o matrícula de la aeronave.

'cn_ln': Esta columna representa el número de construcción y el número de línea de la aeronave. Estos números identifican de manera única una aeronave en particular.

'all_aboard': Esta columna indica el número total de personas a bordo de la aeronave, incluyendo pasajeros y tripulación.

'PASAJEROS A BORDO': Esta columna registra el número de pasajeros a bordo de la aeronave.

'crew_aboard': Esta columna indica el número de miembros de la tripulación a bordo de la aeronave.

'cantidad de fallecidos': Esta columna registra el número total de personas fallecidas en el accidente, incluyendo pasajeros y tripulación.

'passenger_fatalities': Esta columna indica el número de pasajeros fallecidos en el accidente.

'crew_fatalities': Esta columna registra el número de miembros de la tripulación fallecidos en el accidente.

'ground': muertes en tierra.

'summary': Esta columna proporciona un resumen o descripción del accidente.

```python
#Normalizar fecha y crear columnas de mes y año
df['fecha'] = pd.to_datetime(df['fecha'])
df['año'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month
```

```python
#Cambiamos a numericos valores que estan como str 
columns_to_transform = ['all_aboard', 'PASAJEROS A BORDO', 'crew_aboard', 'cantidad de fallecidos', 'passenger_fatalities', 'crew_fatalities','ground',"año", "mes"]

for column in columns_to_transform:
    df[column] = pd.to_numeric(df[column], errors='coerce')
```

```python
import pandas as pd

# Definir una función para normalizar la hora a formato "HH:MM"
def normalize_hour(hour):
    # Eliminar posibles espacios en blanco alrededor de la hora
    hour = hour.strip()
    
    # Obtener la longitud de la hora
    hour_length = len(hour)
    
    # Verificar el formato de la hora y normalizarlo
    if hour_length == 4:  # Formato "HHMM"
        hour = hour[:2] + ':' + hour[2:]
    elif hour_length == 5:  # Formato "HMM" o "HH:MM"
        if ':' not in hour:
            hour = hour[:1] + ':' + hour[1:]
    
    return hour

# Aplicar la función de normalización a la columna 'HORA declarada'
df['HORA declarada'] = df['HORA declarada'].apply(normalize_hour)

# Imprimir el DataFrame con la columna 'HORA declarada' normalizada
print(df['HORA declarada'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Obtener los 5 valores más repetidos de la columna 'route'
top_routes = df['route'].value_counts().head(5).index

# Filtrar el DataFrame original solo para los registros con las rutas más repetidas
df_top_routes = df[df['route'].isin(top_routes)]

# Generar el diagrama de caja
plt.figure(figsize=(10, 8))
sns.boxplot(x='route', y='cantidad de fallecidos', data=df_top_routes)
plt.title('Diagrama de Caja - Cantidad de Fallecidos por Ruta')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Obtener los 5 valores más repetidos de la columna 'route'
top_routes = df['ac_type'].value_counts().head(10).index

# Filtrar el DataFrame original solo para los registros con las rutas más repetidas
df_top_routes = df[df['ac_type'].isin(top_routes)]

# Generar el diagrama de caja
plt.figure(figsize=(20, 18))
sns.boxplot(x='ac_type', y='cantidad de fallecidos', data=df_top_routes)
plt.title('Diagrama de Caja - Cantidad de Fallecidos por tipo de avion')
plt.show()
```

Hay aviones que apesar de su menor cantidad de accidente tienen mayor cantidad de fallecidos, esto se puede dar quiza porque llevavan mas pasajeros. 


```python
import matplotlib.pyplot as plt

# Filtrar y agrupar los datos
df_filtered = df[['ac_type', 'PASAJEROS A BORDO']]
df_grouped = df_filtered.groupby('ac_type')['PASAJEROS A BORDO'].mean().reset_index()

# Ordenar y seleccionar los 10 primeros tipos de avión
df_top_10 = df_grouped.sort_values(by='PASAJEROS A BORDO', ascending=False).head(10)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.bar(df_top_10['ac_type'], df_top_10['PASAJEROS A BORDO'])
plt.xlabel('Tipo de Avión')
plt.ylabel('Promedio de Pasajeros')
plt.title('Pasajeros Promedio por Tipo de Avión (Top 10)')
plt.xticks(rotation=90)
plt.show()

```

Los aviones con mayores pasajeros no son los aviones con mayor accidentes y muertos

```python
import pandas as pd
import matplotlib.pyplot as plt

#Identificamos los 10 aviones mas repetidos
top_10_aviones = df['ac_type'].value_counts().head(10).index


# filtramos el df
df_top_10 = df[df['ac_type'].isin(top_10_aviones)]


# Calcula la cantidad de pasajeros promedio para cada tipo de avión:
promedio_pasajeros = df_top_10.groupby(['ac_type'])['PASAJEROS A BORDO'].mean()
promedio_pasajeros
```
